```
import os.path
from os import path
import operator

#https://github.com/mertemin/turkish-word-list/blob/master/words.txt

def LoadDictionary():
  words=[]
  with open("words.txt") as openfile:
    for line in openfile:
      words.append(line.strip())
  return words


TurkishDictionary = LoadDictionary()

#print(str(TurkishDictionary))


lettersCount = 29
upper = ['A', 'B',	'C',	'Ç',	'D',	'E',	'F',	'G',	'Ğ',	'H',	'I',	'İ',	'J',	'K',	'L',	'M',	'N',	'O',	'Ö',	'P',	'R',	'S',	'Ş',	'T',	'U',	'Ü',	'V',	'Y',	'Z']
lower = ['a',	'b',	'c',	'ç',	'd',	'e',	'f',	'g',	'ğ',	'h',	'ı',	'i',	'j',	'k',	'l',	'm',	'n',	'o',	'ö',	'p',	'r',	's',	'ş',	't',	'u',	'ü',	'v',	'y',	'z']
def ShowOptionsMenu():
  print("---------------------.----")
  print("Choose An Action:")
  print("(e) to encrypt a file")
  print("(d) to decrypt a file")
  print("(h) to hack :) a a cipher")
  print("(q) to to quit\n")

def GetFileName():
  while True:
    fileName = input("Enter name of file: ")
    if path.exists(fileName):
      return fileName
    else:
      print("No such file found!")

def GetKey():
  key = 0
  while True:
    key_text = input("Enter a key (number) for Cezar cipher: ")
    try:           
      key = int(key_text)
      return key
    except:
      print("That was not a number!")
    
def GetIndex(array, char):
  index = 0
  try:
    index = array.index(char)
    return index    
  except ValueError:
    return -1

def EncryptDecryptFile(inputFile, key):
  file = open(inputFile, 'r') 
  result = ""
  while True:
      char = file.read(1)
      if not char: 
          break 
      charIndex = GetIndex(lower,char)
      if charIndex == -1:
          charIndex = GetIndex(upper,char)
          if charIndex == -1: #not found just add it as it is
            result = result+ char
          else:
            encryptedChar = upper[(charIndex + key) % lettersCount]
            result = result + encryptedChar
      else:
        encryptedChar = lower[(charIndex + key) % lettersCount]
        result = result + encryptedChar      
      
      

  file.close()
  return result

  

def WriteToFile(text, fileName):
  #write result to file
  output = open(fileName, "wt")
  n = output.write(text)
  output.close()


def Hack(inputFile, outputFile):
  values = []
  scores = []
  for key in range(1,lettersCount):
    result = EncryptDecryptFile(inputFile, lettersCount-key)
    values.append(result)
  

  
  #Now we have all possible values 
  #split into words check ditionary for every word and create score

  for i in range(len(values)):
    sum = 0.0
  
    words = values[i].split()
    for j in range(len(words)):
      word = words[j]
      
      if SearchDictionary(word):
        sum = sum + 1
    
    
    scores.append(sum)

  print(str(scores))
  #find index of maximum score
  index,val = max(enumerate(scores), key=operator.itemgetter(1))

  #key should be index+1 because we started key from 1 but array start from 0
  print("Predicted key =" + str(index+1))
  WriteToFile(values[index],outputFile)


def SearchDictionary(word):

  if word.strip().lower() in TurkishDictionary:
    return True
  else:
    return False
  

while True:
  ShowOptionsMenu()
  choice = input("What would you like to do:")
  if choice == "e":
    file = GetFileName()
    key = GetKey()
    result = EncryptDecryptFile(file,key)
    WriteToFile(result, "encrypted.txt")
    print("The result is in the file: encrypted.txt")

      
  elif choice == "d":
    file = GetFileName()
    key = GetKey()
    key = lettersCount - key
    result = EncryptDecryptFile(file,key)
    WriteToFile(result, "decrypted.txt")
    print("The result is in the file: decrypted.txt")
    
  elif choice == "h":
    file = GetFileName()
    Hack(file,"hacked.txt")    
    print("The result is in the file: hacked.txt")
  elif choice == "q":
    break
  else:
    print("No such option. Please choose from the menu")
```
